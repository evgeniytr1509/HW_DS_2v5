```python
# Максимальна ціна
max_price = df["price"].max()

# Мінімальна ціна
min_price = df["price"].min()

# Середня ціна
mean_price = df["price"].mean()

# Медіанна ціна
median_price = df["price"].median()


print("Максимальна ціна:", max_price)
print("Мінімальна ціна:", min_price)
print("Середня ціна:", mean_price)
print("Медіанна ціна:", median_price)
```

```python
# Фільтрація книг, що потрапили до Топ-50 у 2015 році
top_2015_books = df[df["year"] == 2015]
# Знаходження найдорожчої книги серед цього підсету
most_expensive_2015_book = top_2015_books[top_2015_books["price"] == top_2015_books["price"].max()]
print("Найдорожча книга серед Топ-50 у 2015 році:")
most_expensive_2015_book
```

```python
# Фільтрація книг за жанром Fiction та роком 2010
fiction_2010_books = df[(df["genre"] == "fiction") & (df["year"] == 2010)]
num_fiction_2010_books = fiction_2010_books.shape[0]
print("Кількість книг жанру Fiction, що потрапили до Топ-50 у 2010 році:", num_fiction_2010_books)
```

```python
# Фільтрація книг за рейтингом у 2015 році та ціною менше 8 доларів
cheap_2015_books = df[(df["year"] == 2015) & (df["price"] < 8)]
# Сортування за зростанням ціни
sorted_cheap_2015_books = cheap_2015_books.sort_values(by="price")
# Відповідь на питання про останню книгу
last_book = sorted_cheap_2015_books.tail(1)  
#  метод tail(1) для отримання останнього рядка
print("Остання книга у відсортованому списку:")
last_book
```

```python
# Групування даних за жанром та підрахунок максимальної та мінімальної ціни
genre_price_stats = df.groupby("genre")["price"].agg([("Максимальна ціна", "max"), ("Мінімальна ціна", "min")])

# Виведення результату
print("Максимальна та мінімальна ціни для кожного з жанрів:")
print(genre_price_stats)
```

```python
# Групування даних за авторами та підрахунок кількості книг
author_book_counts = df.groupby("author").agg(Кількість_книг=("name", "count")).reset_index()

# Виведення таблиці кількості книг для кожного автора
print("Таблиця кількості книг для кожного автора:")
print(author_book_counts)

# Визначення розмірності таблиці
table_shape = author_book_counts.shape

# Знаходження автора з найбільшою кількістю книг
max_books_author = author_book_counts[author_book_counts["Кількість_книг"] == author_book_counts["Кількість_книг"].max()]

# Виведення результату
print("Розмірність таблиці з кількістю книг для кожного автора:", table_shape)
print("Автор з найбільшою кількістю книг:", max_books_author["author"].iloc[0])
print("Кількість книг цього автора:", max_books_author["Кількість_книг"].iloc[0])
```

```python
# Групування даних за авторами та підрахунок середнього рейтингу
author_avg_rating = df.groupby("author").agg(Середній_рейтинг=("user_rating", "mean")).reset_index()

# Визначення автора з найнижчим середнім рейтингом
min_avg_rating_author = author_avg_rating[author_avg_rating["Середній_рейтинг"] == author_avg_rating["Середній_рейтинг"].min()]

# Виведення результату
print("Автор з найнижчим середнім рейтингом:", min_avg_rating_author["author"].iloc[0])
print("Середній рейтинг цього автора:", min_avg_rating_author["Середній_рейтинг"].iloc[0])
# З'єднання двох датафреймів
merged_data = pd.concat([author_book_counts, author_avg_rating["Середній_рейтинг"]], axis=1)

# Сортування за кількістю книг та середнім рейтингом
sorted_data = merged_data.sort_values(by=["Кількість_книг", "Середній_рейтинг"], ascending=[False, True])

# Виведення результату
print("Автор перший у списку:")
sorted_data.head(1)
```
