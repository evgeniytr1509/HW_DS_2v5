```python
import pandas as pd
import matplotlib.pyplot as plt

# Встановлюємо директиву для побудови графіків всередині Jupyter Notebook
%matplotlib inline

# Функція для обчислення середньої зарплати
def fill_avg_salary(row):
    return row["Зарплата.в.месяц"]

# Створюємо новий стовпець "avg" за допомогою функції fill_avg_salary
df["avg"] = df.apply(fill_avg_salary, axis=1)

# Виводимо перші 5 рядків для перевірки
print(df.head())

# Побудова графіку
plt.figure(figsize=(10, 6))
plt.hist(df["avg"], bins=20, color='skyblue')
plt.title('Розподіл середньої зарплати на місяць')
plt.xlabel('Середня зарплата (грн)')
plt.ylabel('Кількість працівників')
plt.grid(True)
plt.tight_layout()

average_salary = df["avg"].mean()
print("Средняя зарплата: {:.2f} грн".format(average_salary))

# Відображення графіку
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Встановлюємо директиву для побудови графіків всередині Jupyter Notebook
%matplotlib inline

# Побудова графіку
plt.figure(figsize=(10, 6))
plt.hist(df["Зарплата.в.месяц"], bins=20, color='skyblue')
plt.title('Розподіл середньої зарплати на місяць')
plt.xlabel('Середня зарплата (грн)')
plt.ylabel('Кількість працівників')
plt.grid(True)
plt.tight_layout()

average_salary = df["Зарплата.в.месяц"].mean()
print("Средняя зарплата: {:.2f} грн".format(average_salary))

# Выполняем агрегацию данных для минимальной и максимальной зарплаты по должности
aggregated_data = df.groupby("Должность")["Зарплата.в.месяц"].agg(['min', 'max', 'mean'])

# Выводим полученный DataFrame
print(aggregated_data)

# Відображення графіку
plt.show()

# Построение диаграммы средней зарплаты по профессиям
plt.figure(figsize=(12, 6))
plt.bar(aggregated_data.index, aggregated_data["mean"], color='skyblue')
plt.title('Средняя зарплата по профессиям')
plt.xlabel('Профессия')
plt.ylabel('Средняя зарплата (грн)')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()

# Відображення графіку
plt.show()



data = df.dropna(subset=['Язык.программирования'])
data = data['Язык.программирования'].value_counts()
labels = data.index
plt.pie(
    data,
    labels=labels,
    shadow=True,
    explode=[0.15]*len(data),
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35,
)
plt.show()



# Рассчитываем среднюю зарплату и сохраняем её в новом столбце "mean"
df["mean"] = df["Зарплата.в.месяц"].mean()

# Создаем описательную статистику для нового столбца "mean"
description_mean = df["mean"].describe()

# Выводим описательную статистику
print(description_mean)
```
